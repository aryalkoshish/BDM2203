````python

# import data processing libraries
import numpy as np
import pandas as pd

# import data visuaization libraries
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
````

````python
# Q1. Load the flight price dataset and examine its dimensions. How many rows and columns does the dataset have?
playstore_df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')
````

````python
# 1. Checking missing values missing data
playstore_df.isna().sum()
###here wwe could see that there are 1474 missing values in just rating column. Now if we want to analize the data without lceaning itt
#let us say we want to know the average rating of the data we have and if we have 1474 ratings with missig values. It would impact on our result
# and our analysis as well. We can deal this with multiple ways.Two ways to deal with are. 1. Removing the missing values 2. Filling the missing places
````

````python
# Remove rows with missing values
cleaned_df = playstore_df.dropna()
cleaned_df
````

````python
cleaned_df['Category'].unique()
````

````python
cleaned_df['Category'].nunique()
````

````python
# average rating of categories
cleaned_df.groupby(by='Category')['Rating'].mean()
````

````python
cleaned_df['Rating'] = cleaned_df['Rating'].fillna(cleaned_df.groupby('Category')['Rating'].transform('mean'))

# Categories with highest average rating (top 5)
rating = pd.DataFrame(cleaned_df.groupby('Category')['Rating'].mean().sort_values(ascending=False)[:5]).reset_index()
rating.columns = ['Category', 'Mean Rating']

plt.figure(figsize=(10.5, 4))
ax = sns.barplot(data=rating, x='Category', y='Mean Rating', palette='Spectral')

ax.set_title('Top 5 rated categories', fontdict={'weight': 'bold', 'color': 'black'})
plt.xticks(rotation=15)

for i in ax.containers:
    ax.bar_label(i,)

plt.show()
````

**What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship**

````python
playstore_df.info()
````

````python
playstore_df['Size'].unique()
````

````python
##Converting teh size of app to the numeric value to make the calculation easier
def convert_size(size):
  if 'M' in size:
    return int(float(size.replace('M',''))*1000)
  elif 'k' in size:
    return float(size.replace('k',''))
  elif 'Varies with device' in size:
    return np.nan
  else:
    return np.nan
````

````python
playstore_df['Size']=playstore_df['Size'].apply(convert_size)
playstore_df.info()
````

````python
playstore_df['Size'].unique()
````

````python
# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(playstore_df['Size'], playstore_df['Rating'], alpha=0.30)
plt.title('Relationship between App Size and Rating')
plt.xlabel('Size (in MB)')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
````

````python

# renaming 'Price' col with 'Price($)'
playstore_df.rename(columns = {"Price": "Price($)"},
          inplace = True)

# converting str to float
playstore_df['Price($)'] = playstore_df['Price($)'].apply(lambda x : float(x.replace('$', '')))
##Already converted to float
````

````python

# converting str to int
playstore_df['Installs'] = playstore_df['Installs'].apply(lambda x : int(x.split('+')[0].replace(',', '')))
playstore_df.head(3)
````

````python
playstore_df['Total Revenue($)'] = playstore_df['Installs'] * playstore_df['Price($)']

playstore_df.head(3)
````

````python

# top 10 apps with highest amount of revenue generated
rev = pd.DataFrame(playstore_df.groupby('App')['Total Revenue($)'].max().sort_values(ascending=False)[:10]).reset_index()

fig, ax = plt.subplots(figsize=(10, 4))
ax_0 = sns.barplot(data=rev, x='Total Revenue($)', y='App', ax=ax, palette='viridis')

ax_0.set_title('Revenue generated by top 10 Apps', fontdict={'weight': 'bold', 'color': 'grey', 'fontsize':15})
for i in ax_0.containers:
    ax_0.bar_label(i,)
````

**Best Time to lunch the app**

````python
# Adding 2 new col : 'Year', 'Month'

playstore_df['Year'] = playstore_df['Last Updated'].apply(lambda x : x.split(', ')[1])
playstore_df['Month'] = playstore_df['Last Updated'].apply(lambda x : x.split(' ')[0])

playstore_df.head(5)
````

````python
# month of update
update_month = pd.DataFrame(playstore_df['Month'].value_counts()).reset_index()
update_month.columns = ['Month', 'Updates']
update_month
````

````python
# Define the order for the months
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
          'August', 'September', 'October', 'November', 'December']

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 4))  # Use only one subplot

# Bar plot
ax = sns.barplot(data=update_month, x='Updates', y='Month', order=months)
ax.set_title('Month wise no. of updates')

# Add labels to the bars
for i in ax.containers:
    ax.bar_label(i)

plt.show()
````
